```python
#Given list of names
names = [" Alice ", "bob", " CHARLIE "] 
#Creating EmptyList To store the Standard names after cleaning
Clean_Standard_names=[]
for i in names:
    #Removing Extra Spaces and Converting all the names in the list to Lowercase
   name=i.strip().lower()
    #Appending the values present in the name to the clean standard names list
   Clean_Standard_names.append(name)
print("Original names in the List:",names)
print("Clean and Standard names after removing spaces and after converting into lower case are:",Clean_Standard_names)
```
