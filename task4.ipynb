```python
#Given input list
messages = ["Hi", "Welcome to the platform", "OK"]
#Checking every element present in the List
for i in messages:
    #Initializing length variable and printing the message
    l=len(i)
    print("Message:",i)
    print("Length:",l)
    #checking the length of message and printing the message accordingly
    if(l>10):
        print("Long Message greater than 10 characters. ")
    else:
        print("Short message.")
    
```
