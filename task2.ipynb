```python
#Given Input List
marks = [45, 78, 90, 33, 60] 
#Intializing Count variable i.e Pass and  Fail to 0
Pass=0
Fail=0
#Checking the condition 
for i in marks:
    if (i>=50):
        Pass=Pass+1
    else:
        Fail=Fail+1
print("Total no.of students:",len(marks))
print("Total no of students passed:",Pass)
print("Total no of students failed:",Fail)
```
